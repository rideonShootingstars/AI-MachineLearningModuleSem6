```python
def sum(lst):
  """
    Calculates the sum of a list of numbers. """
  total = 0
  for num in lst:
    total += num
  return total

def average(lst):
  """
    Calculates the average of a list of numbers. """
  return sum(lst) / len(lst)

def maximum(lst):
  """
    Finds the maximum value in a list of numbers. """
  max_value = lst[0]
  for num in lst:
    if num > max:
      max_value = num
  return max_value

def minimum(lst):
  """
    Finds the minimum value in a list of numbers. """
  min_value = lst[0]
  for num in lst:
    if num < min_value:
      min_value = num
  return min_value



def main():
  print("Choose an operation.\n1.Sum\n2.Average\n3.Maximum\n4.minimum")
  choice = int(input("Enter your choice (1-3): "))
  if(choice <= 0 or choice >4):
    print("Invalid choice")
    return
  while True:
    user_input = input("Enter a list of numbers separated by spaces")

    if not user_input.strip():
      print("Error: Input is empty. Please enter some numbers.")
      continue

    try:
      numbers = [float(item) for item in user_input.split()]
      break

    except ValueError:
      print("Error: Please ensure you only enter numbers separated by spaces.")
  if choice == 1:
    print(f"Sum: {sum(numbers)}")
  if choice == 2:
    print(f"Average: {average(numbers)}")
  if choice == 3:
    print(f"Maximum: {maximum(numbers)}")
  if choice == 4:
    print(f"Minimum: {minimum(numbers)}")
  return
if __name__ == "__main__":
    main()
```

```python
#4.2 List Manipulation

my_list = [1, 2, 3, 4, 5, 6]
def extract_every_other(lst):
    return lst[::2]

print(extract_every_other(my_list))

def slice_List(lst, start, end):
    return lst[start:end]

print(slice_List(my_list, 1, 4))

def reverse_list(lst):
  return lst[::-1]

print(reverse_list(my_list))

def remove_first_last(lst):
  return lst[1:-1]

print(remove_first_last(my_list))

def get_first_n(lst, n):
  return lst[:n]

print(get_first_n(my_list, 3))

def get_last_n(lst,n):
  return lst[-n:]

print(get_last_n(my_list, 4))

def reverse_skip(lst):
  return lst[-2::-2]

print(reverse_skip(my_list))
```

```python
#4.3.1
nested_list = [[1, 2], [3, 4], [5]]

def flatten_list(lst):
  flattened = []
  for sublist in lst:
    flattened.extend(sublist)
  return flattened

print(flatten_list(nested_list))
```

```python
#4.3.3
nested_list = [[1, 2], [3, [4, 5]], 6]

def sum_nested_list(lst):
  total = 0
  for item in lst:
    if isinstance(item, list):
      total += sum_nested_list(item)
    else:
      total += item
  return total


print(sum_nested_list(nested_list))
```

```python
#4.3.4
nested_list = [[1, 2], [3, 2], [4, 5]]
def remove_element(lst, elem):
  for item in lst:
    if isinstance(item, list):
      remove_element(item,elem)
    else:
      if item == elem:
        lst.remove(elem)
  return lst

print(remove_element(nested_list,2))
```

```python
#4.3.5
nested_list = [[1, 2], [3, [4, 5]], 6]

def find_max(lst):
  max = 0
  for item in lst:
    if isinstance(item, list):
      max = find_max(item)
    else:
      if item > max:
        max = item
  return max

print(find_max(nested_list))
```

```python
#4.3.6
nested_list = [[1, 2], [2, 3], [2, 4]]

def count_occurances(lst, elem):
  count = 0
  for item in lst:
    if isinstance(item, list):
      count += count_occurances(item, elem)
    else:
      if item == elem:
        count += 1
  return count

print(count_occurances(nested_list, 2))
```

```python
#4.3.7
nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

def deep_flatted(lst):
  flattened = []
  for item in lst:
    if isinstance(item, list):
      flattened.extend(deep_flatted(item))
    else:
      flattened.append(item)
  return flattened

print(deep_flatted(nested_list))
```

```python
#4.3.8
nested_list = [[1, 2], [3, 4], [5, 6]]

def average_nested(lst):
  sum = 0
  length = 0
  sum += sum_nested_list(lst)
  for item in lst:
    if isinstance(item, list):
      length += len(item)
    else:
      length +=1
  return sum/length

print(average_nested(nested_list))
```
